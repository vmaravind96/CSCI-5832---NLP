```python
# All imports here
import numpy as np
```

```python
# All paths here
train_path = "/content/drive/MyDrive/CSCI 5832/HW3/S21-gene-train.txt"
```

```python
# Load the data
def get_raw_data(path):
  with open(path, "r") as fp:
    raw_data = fp.readlines()
    
  return raw_data
```

```python
# Helper method to process raw data read from file
def get_processed_data(path):

  raw_data = get_raw_data(path)

  # Initialize the necessary variables
  processed = []
  sentence = []
  tags = []

  for line in raw_data:
    stripped_line = line.strip()
    if len(stripped_line):
      _, word, tag = line.strip().split("\t")
      sentence.append(word)
      tags.append(tag)
    else:
      processed.append([sentence, tags])
      sentence, tags = [], []
  
  return processed
```

```python
# Process the train data to get np array in desired format
train_data = get_processed_data(train_path) 
```
